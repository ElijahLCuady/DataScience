```python

# Create a boolean mask to identify rows with missing values
missing_rows_mask = df.isna().any(axis=1)

# Use the mask to filter and print out rows with missing values
missing_rows = df[missing_rows_mask]

# Print the rows with missing values
print(missing_rows)
```

```python
print(df.count())
```

```python
# Filter for job titles containing "developer"
df_developers = df[df["job_title"].str.contains("Software")]

print(df_developers.count())
# Count the rows in the filtered DataFrame
developer_count = df_developers.shape[0]  # Get the number of rows

print(f"Number of job titles containing 'developer': {developer_count}")
```
